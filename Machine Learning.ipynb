##### Melakukan one hot enconding pada kolom type_of_meal_plan, room_type_reserved, dan market_segment_type dikarenakan pada soal tidak diberitahu secara spesifik apakah ada hierarki pada kolom tersebut, maka one hot encoding akan lebih tepat digunakan pada kolom tersebut. Lalu pada kolom arrival_year akan dilakukan label encoding dikarenakan hanya ada 2 unique value pada kolom tersebut.

```python
enc = preprocessing.OneHotEncoder()
def encode_columns(x_train, x_test, columns_to_encode):
    train_enc_list = []
    test_enc_list = []
    
    for column in columns_to_encode:
        train_enc = pd.DataFrame(enc.fit_transform(x_train[[column]]).toarray(), columns=enc.get_feature_names_out(input_features=[column]))
        test_enc = pd.DataFrame(enc.transform(x_test[[column]]).toarray(), columns=enc.get_feature_names_out(input_features=[column]))
        
        train_enc_list.append(train_enc)
        test_enc_list.append(test_enc)
    
    x_train = x_train.reset_index()
    x_test = x_test.reset_index()
    
    for train_enc, test_enc in zip(train_enc_list, test_enc_list):
        x_train = pd.concat([x_train, train_enc], axis=1)
        x_test = pd.concat([x_test, test_enc], axis=1)
    
    for column in columns_to_encode:
        x_train = x_train.drop(column, axis=1)
        x_test = x_test.drop(column, axis=1)
    
    return x_train, x_test

columns_to_encode = ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']

x_train, x_test = encode_columns(x_train, x_test, columns_to_encode)
```

```python
x_train['arrival_year'] = label_encoding.fit_transform(x_train['arrival_year'])
x_test['arrival_year'] = label_encoding.fit_transform(x_test['arrival_year'])
```

##### Menghapus kolom index karena tidak akan digunakan untuk modeling.

##### Dikarenakan pemodelan pada kali ini menggunakan Random Forest dan XGBoost, maka scalling tidak akan dilakukan pada dataset ini. Hal tersebut dikarenakan Random Forest dan XGBoost bisa mengatasi outlier yang ada pada data.

##### Dilakukan reindex untuk kolom x_test dan juga x_train, hal tersebut dilakukan dikarenakan ada beberapa data yang inbalance, dengan melakukan hal ini maka banyaknya kolom pada x_test dan x_train sudah pasti akan sama, sehingga tidak akan ada error yang akan terjadi nantinya.

```python
x_test = x_test.reindex(columns=x_train.columns, fill_value=0)
```

##### Melakukan modeling untuk Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')

RF_class = RandomForestClassifier(random_state = 42)
RF_class.fit(x_train, y_train)
```

```python
y_predict_rf = RF_class.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print('\nClassification Report\n')
print(classification_report(y_test, y_predict_rf, target_names= ['Canceled','Not_Canceled']))
```

```python
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')

XGB_class = XGBClassifier(random_state = 42)
XGB_class.fit(x_train, y_train)
```

```python
y_predict_xgb = XGB_class.predict(x_test)
```

##### Membuat classification report dari model yang telah dibuat

```python
from sklearn.metrics import classification_report
print('\nClassification Report\n')
print(classification_report(y_test, y_predict_xgb, target_names= ['Canceled','Not_Canceled']))
```
